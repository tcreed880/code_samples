```python
from random import randint
from time import process_time

# print the array to verify correctness
def printArray(arr):
    print(arr)

# bubble sort algorithm to sort the array in ascending order
def bubblesort(arr):
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            # swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp

# create a list of 15,000 random integers between 1 and 8000
arr = [randint(1, 8000) for i in range(15000)]

# start the CPU timer before sorting
start_time = process_time()

# use bubble sort
bubblesort(arr)

# stop the CPU timer after sorting
end_time = process_time()

# get the CPU time taken by the sorting process
clock = end_time - start_time

```

```python
import matplotlib.pyplot as plt

# timer dictionary for input sizes and time taken
timer_dict = {
    20000: 28.43,
    15000: 17.61,
    10000: 8.06,
    8000: 5.39,
    6000: 2.57,
    4000: 1.11,
    2000: 0.29
}

# create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(timer_dict.keys(), timer_dict.values(), marker='o', color='blue')

plt.xlabel("Array Size (n)", fontsize=12)
plt.ylabel("Execution Time (seconds)", fontsize=12)
plt.title("Bubble Sort Performance vs. Input Size", fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
